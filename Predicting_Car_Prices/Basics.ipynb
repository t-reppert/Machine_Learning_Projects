```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
cars = pd.read_csv("imports-85.data", names=['symboling','normalized_losses','make','fuel_type', 'aspiration', 'num_doors','body_style','drive_wheels','engine_location','wheel_base','length','width','height','curb_weight','engine_type','num_cylinders','engine_size','fuel_system','bore','stroke','compression_ratio','horsepower','peak_rpm','city_mpg','highway_mpg','price'])
```

```python
cars.info()
```

```python
cars
```

```python
cars.replace('?',np.nan, inplace=True)
```

```python
cars['bore'] = cars['bore'].astype('float')
cars['stroke'] = cars['stroke'].astype('float')
cars['horsepower'] = cars['horsepower'].astype('float')
cars['peak_rpm'] = cars['peak_rpm'].astype('float')
cars['price'] = cars['price'].astype('float')
cars['normalized_losses'] = cars['normalized_losses'].astype('float')
```

```python
cars.drop(['normalized_losses'], axis=1, inplace=True)
cars.info()
```

```python
cars.dropna(axis=0, inplace=True)
```

```python
cars.info()
```

```python
feature_columns = ['symboling', 'wheel_base', 'length', 'width', 'height', 'curb_weight','engine_size','bore','stroke','compression_ratio','horsepower','peak_rpm','city_mpg','highway_mpg']
```

```python
def normalize(df):
    result = df.copy()
    for name in feature_columns:
        max_val = df[name].max()
        min_val = df[name].min()
        result[name] = (df[name] - min_val)/(max_val - min_val)
    return result
norm_cars = normalize(cars)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
# 144
# 49
def knn_train_test(train_col, target_col, df):
    knn = KNeighborsRegressor()
    train_df = df.iloc[:144]
    test_df = df.iloc[144:]
    train_features = train_df[[train_col]]
    train_target = train_df['price']
    knn.fit(train_features, train_target)
    predictions = knn.predict(test_df[[train_col]])
    mse = mean_squared_error(predictions, test_df['price'])
    rmse = np.sqrt(mse)
    return rmse
```

```python
for col in feature_columns:
    print("Column: {col}".format(col=col))
    rmse = knn_train_test(col, 'price', norm_cars)
    print("RMSE: {rmse}".format(rmse=rmse))
```

Lowest RMSE was curb_weight using the default k value

```python
def knn_train_test(train_col, target_col, df, kval=5):
    knn = KNeighborsRegressor(n_neighbors=kval)
    train_df = df.iloc[:144]
    test_df = df.iloc[144:]
    train_features = train_df[[train_col]]
    train_target = train_df['price']
    knn.fit(train_features, train_target)
    predictions = knn.predict(test_df[[train_col]])
    mse = mean_squared_error(predictions, test_df['price'])
    rmse = np.sqrt(mse)
    return rmse
```

```python
k_map = []
feature_dict = {}
for col in feature_columns:
    #print("Column: {col}".format(col=col))
    for k in [1,3,5,7,9]:
        rmse = knn_train_test(col, 'price', norm_cars, kval=k)
        #print("RMSE: {rmse}".format(rmse=rmse))
        k_map.append((k , rmse))
        if col in feature_dict:
            feature_dict[col].append(rmse)
        else:
            feature_dict[col] = [rmse]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(12,10))
x = [ x for x,y in k_map ]
y = [ y for x,y in k_map ]
plt.scatter(x, y)
```

```python
feature_rmse = {}
for feature, rmses in feature_dict.items():
    #print(feature + "\t\t" + str(round(min(rmses),2)))
    feature_rmse[feature] = round(min(rmses),2)
for f,rmse in sorted(feature_rmse.items(),key=lambda x:x[1]):
    print(f, rmse)
```

```python
def knn_train_test(train_cols, target_col, df, kval=5):
    knn = KNeighborsRegressor(n_neighbors=kval)
    train_df = df.iloc[:144]
    test_df = df.iloc[144:]
    train_features = train_df[train_cols]
    train_target = train_df['price']
    knn.fit(train_features, train_target)
    predictions = knn.predict(test_df[train_cols])
    mse = mean_squared_error(predictions, test_df['price'])
    rmse = np.sqrt(mse)
    return rmse
```
